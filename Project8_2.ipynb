**Auto-Encoder**:
It is a neural network that acquires features from input data. It is classified as unsupervised learning, which does not use any supervised data and trains the output layer to produce the same values as the input data.

**RBM**:
Neurons in the visible layer can only bind to neurons in the hidden layer.

**Auto-Encoder => One-way**
*   Learns to make outputs closer to inputsMinimize errors and reconstructions.
*   Intuitive to understand, easy to implement, and easy to reason with



**RBM => Bidirectional**
*   It is mainly useful for classification, regression, dimensionality reduction, etc.
*   It is generative.


**In other words, unlike an autoencoder, which only discriminates one data vector, RBM can generate new data with a given joint distribution.**
